## The function to review

Your classmate says: *"This function reads a list of temperatures and
tells you how many are above a limit."*

```python
def temps(data, x=100):
    # check temps
    r = []
    for t in data:
        if t > x:
            r.append(t)
    print("found " + str(len(r)))
    return r
```

```python
# You can run it to see how it behaves
result = temps([98, 101, 99, 105, 97], 100)
print(result)
```

**Naming issues:**
- `temps` ã doesn't say what the function *does* (maybe `find_temperatures_above_limit`?)
- `data` ã could be `temperatures` to be more specific
- `x` ã could be `limit` or `threshold`
- `r` ã could be `above_limit` or `high_temps`
- `t` ã could be `temperature`

**Documentation:**
- No docstring ã what does `x=100` mean? Fahrenheit? Celsius? What's the default for?
- The comment `# check temps` doesn't add useful information

**Other issues:**
- `print()` inside the function ã this is a side effect. If someone calls this function 1000 times in a loop, they'll get 1000 printed lines. Better to let the caller decide whether to print.
- No check for empty input ã works fine, but the docstring should mention the behavior

**What's done well:**
- The default value for `x` is a nice touch ã makes it easy to call
- Returns a list (not just a count), which is more flexible
- Simple and easy to understand logic

## Docstring Reminder

A Google-style docstring looks like this:

```python
def function_name(param1: type, param2: type) -> return_type:
    """
    One sentence saying what the function does.

    Args:
        param1 (type): What this parameter is.
        param2 (type): What this parameter is.

    Returns:
        return_type: What comes back.

    Example:
        >>> function_name(1, 2)
        3
    """
```

# Grade Calculator

A simple command-line tool that takes a list of student grades and prints
summary statistics including the average, highest, lowest, and how many
students passed.

## Installation

```bash
git clone https://github.com/username/grade-calculator.git
cd grade-calculator
```

Requires: Python 3.8 or later. No external libraries needed.

## Usage

```bash
python grade_calculator.py 85 92 78 61 45
```

**Output:**
```
Average: 72.2
Highest: 92
Lowest: 45
Passed: 4/5
```

A grade of 60 or above counts as passing.
